```python
#1. Histogram :The histogram represents the frequency of occurrence of specific phenomena which lie within a specific range of values and arranged in consecutive and fixed intervals.
import pandas as pd
import matplotlib.pyplot as plt
data =[['AAA', 99, 100, 30000, 'Banglore'], 
       ['BBB', 26, 70, 12000, 'HYD'],
       ['CCC', 32, 60, 45000, 'Chennai'],
       ['DDD', 44, 78, 99000, 'Mumbai'],
       ['EEE', 21, 30, 34000, 'Kochi']
      ]
df = pd.DataFrame(data, columns=['Name', 'Age', 'Marks', 'Salary', 'Address'])
df.hist()
plt.show()
```

```python
#2. Column Chart :A column chart is used to show a comparison among different attributes, or it can show a comparison of items over time.
import pandas as pd
import matplotlib.pyplot as plt
data =[['AAA', 99, 100, 200, 'Banglore'], 
       ['BBB', 26, 70, 120, 'HYD'],
       ['CCC', 32, 60, 150, 'Chennai'],
       ['DDD', 44, 78, 230, 'Mumbai'],
       ['EEE', 21, 30, 240, 'Kochi']
      ]
df = pd.DataFrame(data, columns=['Name', 'Age', 'Marks', 'Salary', 'Address'])
df.plot.bar()
# plot between 2 attributes 
plt.bar(df['Age'], df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
# 3. Box plot chart : A box plot is a graphical representation of statistical data based on the minimum, first quartile, median, third quartile, and maximum.
import pandas as pd
import matplotlib.pyplot as plt
data =[['AAA', 99, 100, 30000, 'Banglore'], 
       ['BBB', 26, 70, 12000, 'HYD'],
       ['CCC', 32, 60, 45000, 'Chennai'],
       ['DDD', 44, 78, 99000, 'Mumbai'],
       ['EEE', 21, 30, 34000, 'Kochi']
      ]
df = pd.DataFrame(data, columns=['Name', 'Age', 'Marks', 'Salary', 'Address'])
df.plot.box()
plt.boxplot(df['Salary'])
plt.show()
```

```python
#4. Pie Chart : A pie chart shows a static number and how categories represent part of a whole the composition of something. A pie chart represents numbers in percentages, and the total sum of all segments needs to equal 100%.
import pandas as pd
import matplotlib.pyplot as plt
data =[['AAA', 99, 100, 30000, 'Banglore'], 
       ['BBB', 26, 70, 12000, 'HYD'],
       ['CCC', 32, 60, 45000, 'Chennai'],
       ['DDD', 44, 78, 99000, 'Mumbai'],
       ['EEE', 21, 30, 34000, 'Kochi']
      ]
df = pd.DataFrame(data, columns=['Name', 'Age', 'Marks', 'Salary', 'Address'])
plt.pie(df['Age'], labels={'A','B','C','D','E'}, autopct='%1.1f%%', shadow=True)
plt.show()
plt.pie(df['Marks'], labels={'A','B','C','D','E'}, autopct='%1.1f%%', shadow=True)
plt.show()
plt.pie(df['Salary'], labels={'A','B','C','D','E'}, autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
#5. Scatter plot : A scatter chart shows the relationship between two different variables and it can reveal the distribution trends. 
import pandas as pd
import matplotlib.pyplot as plt
data =[['AAA', 99, 100, 200, 'Banglore'], 
       ['BBB', 26, 70, 120, 'HYD'],
       ['CCC', 32, 60, 150, 'Chennai'],
       ['DDD', 44, 78, 230, 'Mumbai'],
       ['EEE', 21, 30, 240, 'Kochi']
      ]
df = pd.DataFrame(data, columns=['Name', 'Age', 'Marks', 'Salary', 'Address'])
plt.scatter(df['Age'],df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
