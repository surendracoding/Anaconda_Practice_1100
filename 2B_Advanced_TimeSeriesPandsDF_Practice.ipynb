# 14-video:  Pandas Time Series Analysis-  DatetimeIndex and Resample

```python
import pandas as pd
```

```python
#1). rad csv 
df = pd.read_csv("File5_AppleStock.csv")
print(type(df.date[0]))
df.head(5)
```

```python
#2). parsing date columns within read_csv funtion
df = pd.read_csv("File5_AppleStock.csv", parse_dates=['date'])
print(type(df.date[0]))
df.head(5)
```

```python
#3). change indexes in df
df = pd.read_csv("File5_AppleStock.csv", parse_dates=['date'], index_col = 'date')
print(df.index)
df.head(5)
```

```python
#4). retrie data based on index [select jan-2019 month data]
df['2019-01']
```

```python
#5) print only volume of jan-2019 month
print(df['2019-01'].volume)
```

```python
#6) print only volume aggregation of jan-2019 month
print(df['2019-01'].volume.mean())
#sum(), mean(), max(), min(), 
```

```python
#7) printing data between 2 dates
df['2019-01-02': '2019-01-04']
```

```python
#8). monthly frequency of the data by using resample
df.volume.resample('M').max()
#M- monthly, Y-Yealy, D-daily, Q-quartly
#sum(), mean(), max(), min(), all aggregation funtions
```

```python
#9) ploting graphs on time-series
%matplotlib inline
df.volume.resample('M').min().plot()
#min(), max(), mean(),sum()
```

```python
#9) ploting graphs on time-series with different types of charts
%matplotlib inline
df.volume.resample('M').min().plot(kind='bar')
#df.volume.resample('M').min().plot(kind='hist')
#min(), max(), mean(),sum()
#kind--> pie, hist, bar, 
```

```python
#9) ploting graphs on time-series with different types of charts
%matplotlib inline
df.volume.resample('M').min().plot(kind='pie')
#df.volume.resample('M').min().plot(kind='hist')
#min(), max(), mean(),sum()
#kind--> pie, hist, bar, 
```

```python
#1) DF without date columns
df = pd.read_csv("File6_AppleStock_withoutDates.csv")
df.head(5)
```

```python
df.count()
```

```python
#2). adding date column by start, end, frequency
new_col = pd.date_range(start='6/1/2019', end='8/2/2019')
new_col
```

```python
#3). set inndex of above date_range column = total number of rows in df
df.set_index(new_col, inplace=True)
df
```

```python
#3). slicing the date column
df["2019-06-01":"2019-06-10"]
```

```python
#3). slicing the date column and aggregations
%matplotlib inline
df["2019-06-01":"2019-06-10"].volume.plot()
```

```python
#4). Business days only not on weekends

#1) DF without date columns
df = pd.read_csv("File6_AppleStock_withoutDates.csv")
df.head(5)

#2). adding date column by start, end, frequency
new_col = pd.date_range(start='6/1/2019', end='8/28/2019', freq="B")
new_col

#3). set inndex of above date_range column = total number of rows in df
df.set_index(new_col, inplace=True)
df
```

```python
#5). Predict the values in missing days ==> guess the values in weekends

df.asfreq('D', method='pad')
```

```python
#5). Predict the values in missing days ==> guess the values in weekends  ==> if frequency is hour wise then datetype wii become timestamp

df.asfreq('H', method='pad')
```

```python
#6). if we dont know about end date, then give periods count so that it will get those many days timeseries
```

```python
#1) DF without date columns
df = pd.read_csv("File6_AppleStock_withoutDates.csv")
df.head(5)

#2). adding date column by start, end, frequency ; B==> business days
new_col = pd.date_range(start='6/1/2019', periods=63, freq="B")
new_col

#3). set inndex of above date_range column = total number of rows in df
df.set_index(new_col, inplace=True)
df
```
