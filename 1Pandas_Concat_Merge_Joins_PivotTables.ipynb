```python
import pandas as pd
df1=pd.DataFrame(
{
    "City": ["Bang","HYD","Delhi"],
    "Temper": [20,30,40]
}
)
df1
```

```python
import pandas as pd
df2=pd.DataFrame(
{
    "City": ["HYD","Bang","Chennai","Mumbai"],
    "Rain": [5,6,7,8]
}
)
df2
```

```python
df3=pd.merge(df1,df2, on="City")
df3
```

```python
df4=pd.merge(df1,df2, on="City", how="outer")
df4
```

```python
df5=pd.merge(df1,df2, on="City", how="left")
df5
```

```python
df6=pd.merge(df1,df2, on="City", how="right", indicator=True)
df6
```
